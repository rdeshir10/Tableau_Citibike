```python
import os
import pandas as pd
import numpy as np
```

```python
# Specify the folder containing the CSV files
folder_path_2017 = '2017-citibike-tripdata'
folder_path_2018 = '2018-citibike-tripdata'
folder_path_2019 = '2019-citibike-tripdata'
```

```python
column_mapping = {
    'Trip Duration': 'trip_duration',
    'Start Time': 'start_time',
    'Stop Time': 'stop_time',
    'Start Station ID': 'start_station_id',
    'Start Station Name': 'start_station_name',
    'Start Station Latitude': 'start_station_latitude',
    'Start Station Longitude': 'start_station_longitude',
    'End Station ID': 'end_station_id',
    'End Station Name': 'end_station_name',
    'End Station Latitude': 'end_station_latitude',
    'End Station Longitude': 'end_station_longitude',
    'Bike ID': 'bike_id',
    'User Type': 'user_type',
    'Birth Year': 'birth_year',
    'Gender': 'gender',
    'tripduration': 'trip_duration',
    'starttime': 'start_time',
    'stoptime': 'stop_time',
    'start station id': 'start_station_id',
    'start station name': 'start_station_name',
    'start station latitude': 'start_station_latitude',
    'start station longitude': 'start_station_longitude',
    'end station id': 'end_station_id',
    'end station name': 'end_station_name',
    'end station latitude': 'end_station_latitude',
    'end station longitude': 'end_station_longitude',
    'bikeid': 'bike_id',
    'usertype': 'user_type',
    'birth year': 'birth_year',
    'gender': 'gender'}
```

```python
# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path_2017) if file.endswith('.csv')]

# Initialize an empty list to store DataFrames
df_list_17 = []

# Loop through each CSV file, read the first 300 rows, and append to the list
for file in csv_files:
   

    file_path = os.path.join(folder_path_2017, file)
     # Get the total number of rows in the file
    total_rows = sum(1 for row in open(file_path)) - 1  # Subtract 1 for the header

    # Randomly select rows to skip
    rows_to_skip = sorted(np.random.choice(range(1, total_rows + 1), total_rows - 1000, replace=False))
    df_2017 = pd.read_csv(file_path, nrows=400, skiprows=rows_to_skip) # Limit to the first 300 rows
    df_2017.rename(columns=column_mapping, inplace=True)
    df_list_17.append(df_2017)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df_2017 = pd.concat(df_list_17, ignore_index=True)

# Display the combined DataFrame (first few rows)
combined_df_2017.sample(5)
```

```python
# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path_2018) if file.endswith('.csv')]

# Initialize an empty list to store DataFrames
df_list_18 = []

# Loop through each CSV file, read the first 300 rows, and append to the list
for file in csv_files:
   
    file_path = os.path.join(folder_path_2018, file)
     # Get the total number of rows in the file
    total_rows = sum(1 for row in open(file_path)) - 1  # Subtract 1 for the header

    # Randomly select rows to skip
    rows_to_skip = sorted(np.random.choice(range(1, total_rows + 1), total_rows - 1000, replace=False))
    df_2018 = pd.read_csv(file_path, nrows=400, skiprows=rows_to_skip) # Limit to the first 300 rows
    df_2018.rename(columns=column_mapping, inplace=True)
    df_list_18.append(df_2018)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df_2018 = pd.concat(df_list_18, ignore_index=True)

# Display the combined DataFrame (first few rows)
combined_df_2018.sample(5)
```

```python
# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path_2019) if file.endswith('.csv')]

# Initialize an empty list to store DataFrames
df_list_19 = []

# Loop through each CSV file, read the first 300 rows, and append to the list
for file in csv_files:
    
    file_path = os.path.join(folder_path_2019, file)
    total_rows = sum(1 for row in open(file_path)) - 1  

    rows_to_skip = sorted(np.random.choice(range(1, total_rows + 1), total_rows - 1000, replace=False))
    df_2019 = pd.read_csv(file_path, nrows=400,skiprows=rows_to_skip) 
    df_2019.rename(columns=column_mapping, inplace=True)
    df_list_19.append(df_2019)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df_2019 = pd.concat(df_list_19, ignore_index=True)

# Display the combined DataFrame (first few rows)
combined_df_2019.sample(5)
```

```python
# Synchronizing Data Types to avoid clash in Tableau
combined_df_2017['start_station_id'] = combined_df_2017['start_station_id'].astype('float')
combined_df_2017['birth_year'] = combined_df_2017['birth_year'].astype('float')
combined_df_2017['end_station_id'] = combined_df_2017['end_station_id'].astype('float')
combined_df_2018['start_station_id'].astype('float')
combined_df_2018['birth_year'] = combined_df_2018['birth_year'].astype('float')
combined_df_2019['birth_year'] = combined_df_2019['birth_year'].astype('float')
```

```python
# Dropping Null Values
combined_df_2019.dropna(subset=['end_station_id','end_station_name'], inplace=True)
combined_df_2018.dropna(subset=['end_station_id','end_station_name'], inplace=True)
combined_df_2017.dropna(subset=['user_type','birth_year'], inplace=True)
```

```python
# Concatenate (Union) the data into a single DataFrame
combined_df = pd.concat([combined_df_2017, combined_df_2018, combined_df_2019], ignore_index=True)

# Save the combined data to a new Excel file
combined_df.to_excel('Resources/combined.xlsx', index=False)
```

```python
# Reshape the start and end lat/lon into a single column for lat and lon
start_df = combined_df[['start_station_id', 'start_station_name', 'start_station_latitude', 'start_station_longitude']]
start_df = start_df.rename(columns={'start_station_id': 'station_id', 'start_station_name': 'station_name', 
                                    'start_station_latitude': 'latitude', 'start_station_longitude': 'longitude'})
start_df['station_type'] = 'Start'

end_df = combined_df[['end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude']]
end_df = end_df.rename(columns={'end_station_id': 'station_id', 'end_station_name': 'station_name', 
                                'end_station_latitude': 'latitude', 'end_station_longitude': 'longitude'})
end_df['station_type'] = 'End'

# Concatenate the two dataframes (start and end) together
coordinates = pd.concat([start_df, end_df], ignore_index=True)
coordinates
```

```python
unique_stations = coordinates.drop_duplicates(subset='station_id', keep='first')
```
